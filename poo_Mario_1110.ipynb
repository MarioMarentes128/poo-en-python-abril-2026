#### Conceptos de POO (Programación Orientada a Objetos) en Python

# lenguaje Python,  crear la clase "animal" con los atributos "nombre", "raza" crear el contructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor __init__ inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! ❤")

    # Función (método) para correr
    def correr(self):
        print(f"{self.nombre} está corriendo a toda velocidad. 😎")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 👀")

# --- Creación de objetos (Instanciación) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Rocco", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "Chafa")

# --- Uso de los objetos ---

print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 20) # Separador visual

print(f"Perro 2: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 20) # Separador visual

print(f"Perro 3: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()
print("Programa creado por Mario NL = 53")
```
